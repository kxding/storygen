```python
import torch
import torch.nn.functional as F

# 定义输入
query = torch.rand(32, 8, 128, 1280, dtype=torch.float16, device='cuda')
key = torch.rand(32, 8, 128, 1280, dtype=torch.float16, device='cuda')
value = torch.rand(32, 8, 128, 1280, dtype=torch.float16, device='cuda')

# 将query调整为符合要求的维度
query_reshaped = query.view(-1, query.size(-1)).transpose(0, 1)

# 创建MultiheadAttention模块
multihead_attn = torch.nn.MultiheadAttention(embed_dim=1280, num_heads=8)

# 生成attention mask（示例中全为1，即所有位置都被考虑）
attn_mask = torch.ones(128, 128, device='cuda', dtype=torch.bool)
# 执行attention操作并传入attention mask
output = F.scaled_dot_product_attention(query, key, value, attn_mask)
# output, _ = multihead_attn(query_reshaped, key, value, key_padding_mask=attn_mask)

print(output)
```
